```python
# set the age range
def age_groups_creation(row):
    if row["age"] < 30:
        return "20 - 30 "
    elif row["age"] < 40:
        return "30 - 40"
    elif row["age"] < 50:
        return "40 - 50"
    elif row["age"] < 60:
        return "50 - 60"
    else:
        return "60 - 70"
```

```python
data
```

```python
# make a another column of age range
data["Age Ranges"]= data.apply(age_groups_creation, axis=1)
```

feature Engineering

```python
X= data[["age", "annual Salary", "net worth"]]
Y= data[['car purchase amount']]
```

```python
X
```

```python
Y
```

training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

here test size 0.2 means 80% of data is going in training set and 20 % is going in testing set

```python
len(X_test)
```

```python
len(X_train)
```

scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(Y_test,predictions)))
    print("Root mean squared error on model is {}".format(np.sqrt(mean_squared_error(Y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
predictions_linear_regression = lr.predict(X_test)
```

```python
modelresults(predictions_linear_regression)
```

```python
from sklearn.svm import SVR
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C":[0.01, 0.1, 0.5], "kernel": ["linear", "rbf", "poly"], "degree":[2, 3, 4]}
```

```python
from sklearn.model_selection import GridSearchCV
gridsvr =GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train, Y_train)
```

```python
gridsvr.fit(X_train, Y_train.values.ravel())
```

```python
predictions_gridsvr= gridsvr.predict(X_test)
```

```python
modelresults(predictions_gridsvr)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
param_gridrfr = {"max_depth": [5, 10, 15], "n_estimators": [2, 3, 4, 8, 9, 10]}
```
